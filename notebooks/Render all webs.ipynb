```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

d = """
  dc.w 8,6,7,4,5,5,4,7    ;clover
  dc.w 6,8,4,9,5,11,7,12
  dc.w 8,10,9,12,11,11,12,9
  dc.w 10,8,12,7,11,5,9,4,8,6,-1
"""

l = [(int(x)*10) for x in ','.join([x for x in d.split() if ',' in x]).split(',')]
vertex = [(l[i],l[i+1]) for i in range(0,len(l)-3,2)]
#Change the Size of Graph using Figsize
fig = plt.figure(figsize=(10,10))

#Generating a 3D sine wave
ax = plt.axes(projection='3d')

# assigning coordinates 
x = np.array([x for x,y in vertex])
max_y = max([y for x,y in vertex])
y = np.array([max_y - y for x,y in vertex])
ln = len(x)
z = np.concatenate([np.full((ln-2,ln),0),np.full((2,ln),660)])
# creating the visualization
plt.rcParams['grid.color'] = "navy"
ax.plot_wireframe(x, y, z, color ='blue')
ax.view_init(elev=20,vertical_axis="y")
ax.xaxis.set_pane_color("black")
ax.yaxis.set_pane_color("black")
ax.zaxis.set_pane_color("black")
```

```python
#Import libraries
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

OFFSET = 5
projected = lambda x,m: x + OFFSET if x < m else x - OFFSET


def create3DVertices(vertices):
    x_max = max([x for x,_ in vertices])
    y_max = max([y for _,y in vertices])
    x_mid = int(x_max/2)
    y_mid = int(y_max/2)
    
    SCALE_FACTOR = 0.4
    d_vs = []
    for i in range(0,len(vertices)-1):
        x1,y1 = vertices[i]
        d_vs += [(x1,y1,0)]  # start
        d_vs += [((x1*SCALE_FACTOR)+x_mid,(y1*SCALE_FACTOR)+y_mid,60)] # opposite side

        x2,y2 = vertices[i+1]
        d_vs += [((x2*SCALE_FACTOR)+x_mid,(y2*SCALE_FACTOR)+y_mid,60)] # opposite side
        d_vs += [(x2,y2,0)] # opposite side
        d_vs += [(x1,y1,0)] # opposite side
        d_vs += [(x2,y2,0)] # opposite side
    x = np.array([[x for x,y,z in d_vs]])
    y = np.array([[y for x,y,z in d_vs]])
    z = np.array([[z for x,y,z in d_vs]])
    return(x,y,z)

d = """
  dc.w 7,6,5,5,3,5,1,7    ;dumb-bell
  dc.w 1,9,3,11,5,11,7,10
  dc.w 9,10,11,11,13,11,15,9
  dc.w 15,7,13,5,11,5,9,6,7,6,-1
"""

l = [(int(x)*10) for x in ','.join([x for x in d.split() if ',' in x]).split(',')]
vertices = [(l[i],l[i+1]) for i in range(0,len(l)-1,2)]


#Change the Size of Graph using Figsize
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
x,y,z = create3DVertices(vertices)
# creating the visualization
plt.rcParams['grid.color'] = "navy"
ax.plot_wireframe(x, y, z, color ='blue')
ax.view_init(azim=135,elev=30,vertical_axis="y")
ax.xaxis.set_pane_color("black")
ax.yaxis.set_pane_color("black")
ax.zaxis.set_pane_color("black")
plt.savefig('filename.png', transparent=True)
```
