````python
mem_map = {
"ALEARO.LST": (0x01C6,0x3000),
"ALVROM.LST": (0x3000, 0x3FFE),
"ALWELG.LST": (0x9000, 0xA8AF),
"ALSCO2.LST": (0xA8B4, 0xB1B6),
"ALDIS2.LST": (0xB1B6, 0xC7A0),
"ALEXEC.LST": (0xC7A0, 0xCB01),
"ALSOUN.LST": (0xCB01, 0xCDDD),
"ALVROM.LST": (0xCDDD, 0xCF24),
"ALCOIN.LST": (0xCF24, 0xD030),
"ALLANG.LST": (0xD030, 0xD703),
"ALHAR2.LST": (0xD703, 0xD7E1),
"ALTES2.LST": (0xD7E1, 0xDDDC),
"ALEARO.LST": (0xDDDC, 0xDF09),
"ALVGUT.LST": (0xDF09, 0xDFDB),
}
````

````python
try:
    int("LIEN", 16)
except:
    print("LIEN")
````

````python
labels = {}
for lst in mem_map:
    start, end = mem_map[lst]
    lines = open("listings/" + lst, "r").readlines()
    for l in lines:
        try:
            offset = int(l[9:13], 16)
        except:
            continue
        label = l[40:47].strip().replace(":", "")
        if not label:
            continue
        addr = start + offset
        hex_addr = hex(addr)[2:].upper()
        labels[label] = hex_addr
labels
        

    
````
