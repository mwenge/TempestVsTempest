```python
from itertools import groupby
from operator import itemgetter
from matplotlib import pyplot as plt
import os

plt.style.use('dark_background')
plt.rcParams['figure.dpi'] = 150

color_tables = [
    {
        "PDIWHI":"WHITE",
        "PDIRED":"RED",
        "PDIYEL":"YELLOW",
    },
    {
        "PDIWHI":"YELLOW",
        "PDIRED":"WHITE",
        "PDIYEL":"RED",
    },
    {
        "PDIWHI":"RED",
        "PDIRED":"YELLOW",
        "PDIYEL":"WHITE",
    },
    {
        "PDIWHI":"WHITE",
        "PDIRED":"WHITE",
        "PDIYEL":"WHITE",
    },
]

def scvec(old_x, old_y, scale, color_table, color, x, y, i = 0):
    c = "BLACK" if i == 0 else color_table.get(color,color)
    x = x * scale
    y = y * scale
    x = x / 2 # Divide by 2: Hack the x scale to match Tempest
    vec = (x - old_x, y-old_y, c)
    old_x = x
    old_y = y
    return (vec,old_x,old_y)

def parse_vector_commands(vector_data, color_table, default_color="red"):
    scale = 1
    sp_scale = {
        0 : 1,
        1 : .5,
        2 : .25
    }
    sl_scale = {
        0 : 1,
        40 : .5,
    }

    # Read in the vector commands from the Tempest source code
    # and translate them into a series of vectors, i.e. a  list of 
    # tuples of the form (x,y,color).
    # This list can then be transformed into a series of x,y co-ordinates
    # for plotting on a graph.
    color = ""
    vecs = []
    old_x,old_y = 0,0
    CALBRITE = "BLACK"
    for l in vector_data.splitlines():
        if not l.strip():
            continue
        tokens = l.strip().split()
        cmd, parms = tokens[0],tokens[1:]
        if cmd in ["ICVEC","ICALVE"]:
            vecs+=[(0,0,"BLACK")]
            old_x,old_y=0,0
        elif cmd == "CSTAT":
            color = parms[0]
        elif cmd == "SCVEC":
            parms = [int(x.replace('.',''),16) for x in parms[0].split(',')]
            vec,old_x,old_y = scvec(old_x, old_y, scale, color_table, color, *parms)
            vecs+=[vec]
        elif cmd == "SCAL":
            parms = [int(x) for x in parms[0].split(',')]
            SP,SL = parms
            scale = sp_scale[SP] + (sp_scale[SP] * sl_scale[SL])
    return vecs

def vectors_to_coords(vectors):
    # Turns a series of vectors into a series of x,y co-ordinates
    # or vertices.
    xs,ys,cs=[],[],[]
    px,py,pc=0,0,None
    for x,y,c in vectors:
        # If we're changing color, add a vertex with the
        # new color at the old position.
        if pc and c != pc:
            xs+=[px]
            ys+=[py]
            cs+=[c]
        # Calculate the position of the new vertex
        px+=x
        py+=y
        xs+=[px]
        ys+=[py]
        cs+=[c]
        pc=c
    return list(zip(xs,ys,cs))

def plot_vertices(verts,file_name,points="",linewidth=1.2):
    # Plot a series of vertices onto a graph. In other words,
    # create an image from them.
    _, ax = plt.subplots()
    ax.tick_params(axis='both', which='major', labelsize=15)
    plt.xlim(-100,100)
    plt.ylim(-100,100)
    last_x,last_y=[],[]
    for k,g in groupby(verts, key=itemgetter(2)):
        l = list(g)
        xs,ys,_ =zip(*l)
        # Don't paint the vectors marked as black, 
        # we're just moving the beam.
        alpha = 0 if k == "BLACK" else 1
        ax.plot(xs,ys,points+'-',color=k,alpha=alpha,linewidth=linewidth)
    plt.savefig(f"splat/vec_image_{file_name}.png", bbox_inches='tight')

def plot_vertices_imgonly(verts,file_name,points="",linewidth=2.0):
    # Plot a series of vertices onto a graph. In other words,
    # create an image from them.
    _, ax = plt.subplots()
    plt.xlim(-100,100)
    plt.ylim(-100,100)
    last_x,last_y=[],[]
    for k,g in groupby(verts, key=itemgetter(2)):
        l = list(g)
        xs,ys,_ =zip(*l)
        # Don't paint the vectors marked as black, 
        # we're just moving the beam.
        alpha = 0 if k == "BLACK" else 1
        ax.plot(xs,ys,points+'-',color=k,alpha=alpha,linewidth=linewidth)
    plt.axis('off')
    plt.savefig(f"splat/vec_image_{file_name}.png", bbox_inches='tight')

```

```python
splat_vectors = [
f"""
SPLAT1: 
        SCAL 0,0
        JSRL SPLAT
        {splat_data}
""",
f"""
SPLAT2: 
        SCAL 0,40
        JSRL SPLAT
        {splat_data}
""",
f"""
SPLAT3: 
        SCAL 1,0
        JSRL SPLAT
        {splat_data}
""",
f"""
SPLAT4: 
        SCAL 1,40
        JSRL SPLAT
        {splat_data}
""",
f"""
SPLAT5: 
        SCAL 2,0
        JSRL SPLAT
        {splat_data}
""",
f"""
SPLAT6: 
        SCAL 2,40
        JSRL SPLAT
        {splat_data}
""",
]

splat_table = [6,5,4,3,2,1,3,5,6]
for i,s in enumerate(splat_table):
    vd = ''.join(splat_vectors[s-1:])
    for j,c in enumerate(range(1,-1,-1)):
        vectors = parse_vector_commands(vd, color_tables[c])
        vertices = vectors_to_coords(vectors)
        plot_vertices(vertices,f"splat_{i}_{j}")
        plot_vertices_imgonly(vertices,f"img_splat_{i}_{j}", linewidth=1.1)
```
